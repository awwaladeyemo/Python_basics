```python
from skimage import data
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 15 ))
#sample image of dataset from library
coffee = data.coffee()
plt.subplot(1,2,1)
plt.imshow(coffee)
gray_coffee=rgb2gray(coffee)
plt.subplot(1,2,2)
plt.imshow(gray_coffee,cmap='gray')
plt.title('Original Image')
plt.show()
```

```python
#RGB TO HSV (hue,saturation,Value)
from skimage.color import rgb2hsv
plt.figure(figsize=(15,15 ))
coffee=data.coffee()
plt.subplot(1,2,1)
plt.imshow(coffee)
hsv_coffee=rgb2hsv(coffee)
plt.subplot(1,2,2)
hsv_coffee_colorbar=plt.imshow(hsv_coffee)
plt.colorbar(hsv_coffee_colorbar,fraction=0.046,pad=0.04)
```

```python
#supervised Segmentation
from skimage import filters
coffee=data.coffee()
gray_coffee=rgb2gray(coffee)
plt.figure(figsize=(12,12 ))
for i in range(10):
  binarized_gray=(gray_coffee>i*0.1)*1
  plt.subplot(5,2,i+1)
  plt.title("threshold>"+str(round(i*0.1,1)))
  plt.imshow(binarized_gray,cmap='gray')
plt.tight_layout()
```
